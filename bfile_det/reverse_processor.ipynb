```python
import numpy as np
import os
from hilbert_curve import hilbertize_arr
import matplotlib.pyplot as plt
import random
import cv2

with open('g_files.txt', 'r') as f:
    lines = f.read().split('\n')
    # print(lines)

ghash = {}
for l in lines:
    ghash.update({l: True})

b_train = []

fp_train = './additional_bexe'

fp_arr = os.listdir(fp_train)
fp_arr.reverse()

for j in range(336, len(fp_arr)):
    try:
        tmp = ghash[fp_arr[j]]
        fp = fp_train

        bytes = np.array([f"{n:08b}" for n in open(fp + '/' + fp_arr[j], "rb").read()])
        for i in range(len(bytes)):
            bytes[i] = int(bytes[i], 2)

        bytes = bytes.astype(np.uint8)

        arr = hilbertize_arr(bytes)
        b_train.append(arr)

        cv2.imwrite(f'./b_train_extra/file{j}REVERSE.png', arr)
        print(f'[ Processed File {j} => {fp_arr[j]} ]')
        # plt.imshow(arr)
        # break
    except KeyError:
        pass

plt.imshow(b_train[random.randint(0, len(b_train)-1)])
```
